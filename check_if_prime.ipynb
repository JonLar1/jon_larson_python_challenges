# Check if Prime

```python
def is_prime(n):
    if n < 2:
        return False

    for i in range(2, n):
        if n % i == 0:
            return False

    else:
        return True
```

```python
# Test

print(is_prime(2)) 
print(is_prime(11))  
print(is_prime(15))
print(is_prime(1))
print(is_prime(0))
```
