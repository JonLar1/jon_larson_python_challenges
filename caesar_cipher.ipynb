# Caesar Cipher

```python
def caesar_cipher(text, shift):
    result = ""
    
    for char in text:
        if not char.isalpha():
            result += char
            continue
            
        if char.isupper():
            char_code = ord(char)
            position = char_code - 65
            new_position = (position + shift) % 26
            new_char_code = new_position + 65
            new_char = chr(new_char_code)
            result += new_char
            
        else:
            char_code = ord(char)
            position = char_code - 97
            new_position = (position + shift) % 26
            new_char_code = new_position + 97
            new_char = chr(new_char_code)
            result += new_char
            
    return result
            
    


   

    
```

```python
# Test

print(caesar_cipher("abc", 3))
print(caesar_cipher("xyz", 2))
print(caesar_cipher("Hello, World!", 5))
```
