# Reverse A String

```python

# Create the function

def reverse_string(text):
    return text[::-1]
```

```python
print(reverse_string("hello"))
print(reverse_string("Python"))
```

# Bonus Challenge

```python
def reverse_string_loop(text):
    reversed_text = ""
    for char in text:
        reversed_text = char + reversed_text
    
    return reversed_text
    
```

```python
print(reverse_string_loop("hello"))
```
